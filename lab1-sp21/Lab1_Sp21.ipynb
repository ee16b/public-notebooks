We will organize the breadboard in the following way:

- The top and bottom + power rails will be 5V, while the - rails will be gnd.
- Block A35-J38 will be reserved for the opamp chip. We will explain the pinout and wire routing later.
- Column 37F-37J is the pin where the signal from the microcontroller (P3.2) will be connected to. Note that the labelled P3.2 is different from pin 13 on the Arduino.


### Circuit Guidelines
- Make sure you use consistent colors for ground, power voltages, etc. Black is usually reserved for gnd, and red for power.
- For easier readability of your layout, when creating wires it is possible to "bend" them by "clicking their path". Try to avoid crossing wires as much as you can for better visualization.
- In general, it's better to have shorter wires, so that it's easier to follow.
- The + and - rails are reserved for power and gnd, respectively. Avoid connecting any other signals to those rails.
- Pay attention to how the two supply multimeters are connected, and connect the third unconnected meter appropriately to measure the input signal.
- Make sure all voltage supplies have the same reference voltage. They should all have their negative terminals connected to the negative power rail (-).
- DO NOT connect multiple wires to the same hole. Tinkercad lets you do this, but this is physically impossible to do on a real breadboard, so do not do this.
- Make sure your components like resistors go into holes on the breadboard, and DO NOT have any legs dangling in empty space. Tinkercad will let you have floating components, but you can't physically do this with real hardware, so don't do this.

#### Testing your circuit

TinkerCad can simulate an Arduino microprocessor board, which is similar to the MSP430 Launchpad. We will use the Arduino to generate a voltage square wave. You can click on the Code button to see the Scratch code the Arduino will be running. The code will turn on pin 13 for two seconds, and then turn it off for 2 seconds. 

Unlike some other microcontrollers, such as the MSP430 Launchpad, the Arduino has 5V logic, while the MSP430 has 3.3V logic. Fortunately, the Arduino has a 3.3V supply pin. We will use the logic pin 13 to drive a CMOS inverter switch that is powered by this 3.3V supply pin. This way, when pin 13 on the Arduino is high, the CMOS inverter will output 0V. When pin 13 is low, the CMOS inverter will output 3.3V.  This way, our TinkerCad simulation will mimic the Launchpad better. The output of the CMOS inverter is labelled P3.2 to represent P3.2 on a Launchpad.

The CMOS inverter, driven by the Arduino pin 13,  generates a square wave with the following properties:

<center>
<img width='700px'
src="images/sqWave.png">
</center>


    - 0.25 Hz frequency (4 seconds cycle)
    - 50% duty cycle (1/2 time on, 1/2 time on during a cycle)
    - 3.3 volts peak-to-peak (Vpp). Peak-to-peak means the difference between the maximum and the minimum of the wave.
    - 1.65V offset

- Connect the output of the inverter ( _labeled P3.2 in TinkerCad_ ) to the input of your buffer circuit ( _labeled P3.2 in the circuit diagram_ ).
- Connect the ground of the Arduino to the negative power rail (-) (which should be $V_{ss}$), so that the Arduino uses the same voltage reference as the other voltage supplies. 

- The LM741 needs a 5V supply to operate. Connect the correct DC voltages to $V_{\text{dd}}$ and $V_{\text{ss}}$.


- Connect the Multimeter to the input of the circuit (P3.2) to measure its voltage.
- Run the simulation
